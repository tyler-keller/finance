```python
import yfinance as yf 
import numpy as np
import pandas as pd
```

```python
from datetime import datetime

tick = yf.Ticker('SPY')
hist = tick.history(interval='1mo', start='2003-02-17', end='2024-07-27').reset_index()
hist
```

```python
import plotly.express as px
```

```python
px.line(hist, x='Date', y='Open')
```

```python
hist['next_open'] = hist['Open'].shift(-1)
```

```python
hist['open_diff'] =  hist['next_open'] - hist['Open']
```

```python
hist
```

```python
hist['next_day_down'] = hist['open_diff'] < 0
hist
```

```python
class Share():
    def __init__(self, buy_price, sell_price=None):
        self.buy_price = buy_price
        self.sell_price = sell_price
        if self.sell_price and self.buy_price:
            self._get_profit()
    
    def _get_profit(self):
        self.profit = self.sell_price - self.buy_price
    
    def sell(self, sell_price):
        assert not self.sell_price
        self.sell_price = sell_price
        self._get_profit()


class Portfolio():
    def __init__(self, cash, shares: list[Share] = []):
        self.cash = cash
        self.original_cash = cash
        self.shares = shares
        self.profit = 0
    
    def buy_shares(self, n, buy_price):
        for _ in range(n):
            new_share = Share(buy_price=buy_price)
            self.shares.append(new_share)
            self.cash -= buy_price
            
    def get_num_shares(self):
        return len(self.shares)

    def sell_all_shares(self, sell_price):
        if self.shares:
            for i in range(self.get_num_shares()):
                self.shares[i].sell(sell_price=sell_price)
                self.profit += self.shares[i].profit
            self.shares = []


cash_amount = 100_000
portfolio = Portfolio(cash=cash_amount)

sell_dates = []
buy_dates = []
for idx, row in hist.iterrows():
    if row['next_day_down']:
        print(row['next_day_down'], f'SELLING ALL {row['Open']}')
        portfolio.sell_all_shares(sell_price=row['Open'])
    else:
        print(row['next_day_down'], f'BUYING 10 AT {row['Open']}')
        portfolio.buy_shares(n=10, buy_price=row['Open'])

print(portfolio.profit)
print(portfolio.cash)
print(portfolio.original_cash)

fig = px.line(hist, x='Date', y='Open')
fig.add_vline(x=2.5, line_width=3, line_dash="dash", line_color="green")
```
